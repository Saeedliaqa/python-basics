```python
n= 20
prime=True
for i in range(2 ,n // 2 +1):
    if n % i ==0:
        prime = False
        break

if prime == True:
    print("prime num")
else:
    print("not prime")
```

sum of n numbers

```python
n=10
sum=0
i=0
while i <=n: 
    sum=sum+i
    i=i+1
print(sum)
```

prime number between 1 & 100

```python
for i in range(101):
    prime= True
    for j in range(2,i//2+1):
        if i % j ==0:
            prime = False
            break

    if prime == True:
        print(i)
```
